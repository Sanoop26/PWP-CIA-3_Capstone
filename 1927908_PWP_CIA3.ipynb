## Data Visualization: 

```python
# Plotting a pie chart for browsers

plt.rcParams['figure.figsize'] = (18, 7)
size = [10551, 1694, 85]
colors = ['darkorange', 'blue', 'green']
labels = "Returning Visitor", "New_Visitor", "Others"
explode = [0, 0, 0.1]
plt.subplot(1, 2, 1)
plt.pie(size, colors = colors, labels = labels, explode = explode, shadow = True, autopct = '%.2f%%')
plt.title('Different Visitors', fontsize = 25)
plt.axis('off')
plt.legend()

# plotting a pie chart for browsers
size = [7961, 2462, 736, 467,174, 163, 300]
colors = ['brown', 'yellow', 'green', 'crimson', 'lightgreen', 'cyan', 'blue']
labels = "2", "1","4","5","6","10","others"

plt.subplot(1, 2, 2)
plt.pie(size, colors = colors, labels = labels, shadow = True, autopct = '%.2f%%', startangle = 180)
plt.title('Different Browsers', fontsize = 25)
plt.axis('off')
plt.legend()
plt.show()
```

```python
# What region do shoppers come from?

plt.figure(figsize = (16,8))
sns.countplot(x = "Region", data = data, 
              order = data['Region'].value_counts().index)
plt.title("Site Visitors by Region", fontsize = 16)
plt.xlabel("Region", fontsize = 16)
plt.ylabel("Number of Visitors", fontsize = 16)
```

```python
# SpecialDay vs No of Visitor's

plt.figure(figsize = (16,8))
sns.countplot(x = "SpecialDay", data = data, 
              order = data['SpecialDay'].value_counts().index)
plt.title("Site Visitors by SpecialDay", fontsize = 16)
plt.xlabel("SpecialDay", fontsize = 16)
plt.ylabel("Number of Visitors", fontsize = 16)
```

```python
# Different Operating Systems used by Visitor's

plt.figure(figsize = (16,8))
sns.countplot(x = "OperatingSystems", data = data, 
              order = data['OperatingSystems'].value_counts().index)
plt.title("OS used by Visitor's", fontsize = 16)
plt.xlabel("OperatingSystems", fontsize = 16)
plt.ylabel("Number of Visitors", fontsize = 16)
```

```python
# Graph to understand the traffic type

plt.figure(figsize = (16,8))
sns.countplot(x = "TrafficType", data = data, 
              order = data['TrafficType'].value_counts().index)
plt.title("Understanding the Traffic Type", fontsize = 16)
plt.xlabel("TrafficType", fontsize = 16)
plt.ylabel("Number of Visitors", fontsize = 16)
```

```python
# Month wise visitor's

data['Month'].value_counts().plot(kind = "bar")
plt.xticks(rotation = 90)
plt.title("Month wise visitor's", fontsize = 16)
plt.xlabel("Month", fontsize = 16)
plt.ylabel("Number of Visitors", fontsize = 16)
plt.show()
```

```python
# Visitor type vs revenue

data = pd.crosstab(data['VisitorType'], data['Revenue'])
data.plot(kind = 'bar', stacked = True, figsize = (8, 5), color = ['lightgreen', 'green'])
plt.title('Visitor Type vs Revenue')
plt.show()
```
